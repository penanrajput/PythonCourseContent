```python
import matplotlib.pyplot as plt
x = [1,2,3] 
y = [2,4,1] 
plt.title("one line")
plt.xlabel('x - axis') 
plt.ylabel('y - axis') 
plt.plot(x, y, 'b', label="Line 1") 
```

```python
import matplotlib.pyplot as plt

plt.title("two line")
plt.xlabel('x - axis') 
plt.ylabel('y - axis') 

plt.plot([1,2,3], [2,4,1], 'b', label="Line 1") 
plt.plot([1,2,3], [3,1,2], 'b', label="Line 2", color="green")

plt.legend() 
```

```python
import matplotlib.pyplot as plt

plt.title("two line")
plt.xlabel('x - axis') 
plt.ylabel('y - axis') 

plt.plot([1,2,3], [2,4,1], 'b', label="Line 1", color="magenta", marker='o', linestyle='dashed') 
plt.plot([1,2,3], [3,1,2], 'b', label="Line 2", color="green", marker='o', linestyle='dashed')

plt.legend() 
```

```python
import matplotlib.pyplot as plt

plt.title("two line")
plt.xlabel('x - axis') 
plt.ylabel('y - axis') 

plt.plot([1,2,3], [2,4,1], 'b', label="Line 1", color="magenta", marker='o', linestyle='-.') 
plt.plot([1,2,3], [3,1,2], 'b', label="Line 2", color="red",   marker='*', linestyle=':')

plt.legend() 
```
